```python
import numpy as np
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
width, height = 100, 100


f = '../mnist_sequence3_sample_8distortions_9x9.npz'
```

```python
data = np.load(f)

train = data['X_train']
valid = data['X_valid']
test = data['X_test']

print "Train size", train.shape
print "Valid size", valid.shape
print "Test size ", test.shape
```

```python
for i in range(3):
    img = train[i].reshape((height, width))
    plt.imshow(img, cmap='gray', interpolation='none')
    plt.title("Train " + str(i), fontsize=20)
    plt.show()
    
for i in range(3):
    img = valid[i].reshape((height, width))
    plt.imshow(img, cmap='gray', interpolation='none')
    plt.title("Valid " + str(i), fontsize=20)
    plt.show()
    
for i in range(3):
    img = test[i].reshape((height, width))
    plt.imshow(img, cmap='gray', interpolation='none')
    plt.title("Test " + str(i), fontsize=20)
    plt.show()
    
    
```
